### LINEAR REGRESSION
##### This finds the line of best fit to summarize the trend in the data, allowing you to predict future outcomes and understand relationships between variables.
##### Linear Regression is  used to model the relationship between a dependent variable and one or more independent variables


##### The primary aim is to understand and predict the growth of iPhone adoption. We're doing this by exploring the relationship between Year and the No of iPhone Users.

##### Mean Absolute Error(MAE) is commonly used to evaluate the accuracy of regression models in machine learning. It's a straightforward metric that helps you understand how far off, on average, a model's predictions are from the actual values. 

##### Mean Squared Error(MSE) is the average squared difference between the value observed in a statistical study and the values predicted from a model.

##### Co-efficient of Determination (r^2) is a statistical measure indicating the proportion of variance in a dependent variable explained by a regression model's independent variable(s), with values ranging from 0 to 1 and higher values signifying a better model fit.
##### The coefficient of determination (r^2) tells you how well your model explains the variations in a set of data.

##### MAPE stands for Mean Absolute Percentage Error. It is a statistical measure used to evaluate the accuracy of a forecasting or predictive model. It works by measuring the average percentage difference between the actual values and the predicted values. A lower MAPE value indicates a higher level of accuracy, with a MAPE of 0% signifying a perfect prediction. 

```python
model_lr = LinearRegression() # At this stage, the model has been created but has not learned anything from your data. It is an empty container ready to be trained.
model_lr.fit(X, y) # The .fit() method takes your training data and uses it to find the best possible straight-line relationship between the independent variable X and the dependent variable y
# Predictions
y_pred = model_lr.predict(X) #This line uses the model I just trained to generate predictions.
#This step is crucial for evaluating how well the model performed on the data it was trained on. It allows you to compare the y_pred values with the original y values.
```

```python
results = []
```

```python
def evaluate_model(name, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    # Accuracy defined from error ratio
    accuracy = 100 - (np.mean(np.abs((y_true - y_pred) / y_true)) * 100)

    
    return {"Model": name, "MAE": mae, "MSE": mse, "RMSE": rmse, "R2": r2,"Accuracy (%)": accuracy}
```

```python
results.append(evaluate_model("Linear Regression", y_test, lr_preds))
results.append(evaluate_model("Random Forest", y_test, rf_preds))
results.append(evaluate_model("XGBoost", y_test, xgb_preds))
```

```python
# Comparison Table
results_df = pd.DataFrame(results)
print("\nModel Comparison Table:")
results_df
```

```python
metrics_to_plot = ["MAE", "RMSE", "MSE", "R2"]
# Create a numerical range for the x-axis tick locations
x_pos = np.arange(len(results_df["Model"]))
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()
for i, metric in enumerate(metrics_to_plot):
    axes[i].bar(results_df["Model"], results_df[metric], color=['skyblue','lightgreen','salmon'])
    axes[i].set_title(metric)
    axes[i].set_ylabel(metric)
    axes[i].set_xticks(x_pos)
    axes[i].set_xticklabels(results_df["Model"], rotation=30)
     

plt.tight_layout()
plt.show()
```
