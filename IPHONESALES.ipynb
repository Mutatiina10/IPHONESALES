#### Split the data into training,Validation and testing.

```python
# Define the features (X) and target (y)
#  We want to predict 'No_of_iPhone_Sold'
X = df_scaled.drop('No_of_iPhone_Sold', axis=1)
y = df_scaled['No_of_iPhone_Sold']
```

```python
# Perform a time-based split
# Use the first 7 years for training and the last 2 for testing
X_train_time = X.iloc[:-2]
X_test_time = X.iloc[-2:]
y_train_time = y.iloc[:-2]
y_test_time = y.iloc[-2:]
```

```python
print("\nShape of X_train (time-based):", X_train_time.shape)
print("Shape of X_test (time-based):", X_test_time.shape)
print("Shape of y_train (time-based):", y_train_time.shape)
print("Shape of y_test (time-based):", y_test_time.shape)
print("\nFirst 5 rows of X_train (time-based):")
print(X_train_time.head())
```

#### MODEL TRAINING AND EVALUATION

### LINEAR REGRESSION
##### This finds the line of best fit to summarize the trend in the data, allowing you to predict future outcomes and understand relationships between variables.
##### Linear Regression is  used to model the relationship between a dependent variable and one or more independent variables

##### The primary aim is to understand and predict the growth of iPhone adoption. We're doing this by exploring the relationship between Year and the No of iPhone Users.

##### Mean Absolute Error(MAE) is commonly used to evaluate the accuracy of regression models in machine learning. It's a straightforward metric that helps you understand how far off, on average, a model's predictions are from the actual values. 

##### Mean Squared Error(MSE) is the average squared difference between the value observed in a statistical study and the values predicted from a model.

##### Co-efficient of Determination (r^2) is a statistical measure indicating the proportion of variance in a dependent variable explained by a regression model's independent variable(s), with values ranging from 0 to 1 and higher values signifying a better model fit.
##### The coefficient of determination (r^2) tells you how well your model explains the variations in a set of data.

##### MAPE stands for Mean Absolute Percentage Error. It is a statistical measure used to evaluate the accuracy of a forecasting or predictive model. It works by measuring the average percentage difference between the actual values and the predicted values. A lower MAPE value indicates a higher level of accuracy, with a MAPE of 0% signifying a perfect prediction. 

```python
# Define MAPE function
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# Model Training (at least 3 algorithms)
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'XGBoost': xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'SVR': SVR(kernel='rbf', C=1.0, epsilon=0.1)
}
```

```python
metrics = {}
```

```python
for name, model in models.items():
    model.fit(X_train_time, y_train_time)
    y_pred = model.predict(X_test_time)
    metrics[name] = {
        'MSE': mean_squared_error(y_test_time, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_test_time, y_pred)),
        'MAE': mean_absolute_error(y_test_time, y_pred),
        'MAPE': mape(y_test_time, y_pred),
        'R²': r2_score(y_test_time, y_pred)
    }
    print(f"\n{name} - Test MSE: {metrics[name]['MSE']:.4f}, RMSE: {metrics[name]['RMSE']:.4f}, MAE: {metrics[name]['MAE']:.4f}, MAPE: {metrics[name]['MAPE']:.4f}%, R²: {metrics[name]['R²']:.4f}")
```

**JUSTIFICATION**
Linear Regression serves as a solid baseline for this dataset, which involves predicting continuous values like Number of iPhones Sold, MarketShare. It is appropriate for a regression problem, especially as an initial, simple model to establish a performance benchmark for more complex algorithms. Its straightforward nature makes it easy to implement and interpret, providing a clear starting point for understanding the relationships in the data.

Strengths: Has High interpretability and is easy to use.
Weaknesses: The assumption of linearity means it will likely perform poorly if the market data has complex or non-linear patterns. It is also sensitive to outliers and assumes a constant variance of errors.

Assumptions: It assumes a linear relationship between independent and dependent variables.

 **JUSTIFICATION**
 
XGBoost is a highly effective and popular gradient boosting algorithm that is well-suited for high-stakes predictive tasks, including time-series forecasting.

Strengths: It is also highly effective at capturing complex, non-linear patterns.

Weaknesses: Interpreting the final model can also be difficult.

Assumptions: XGBoost makes fewer assumptions than linear regression about the underlying data distribution.
             The model relies on the assumption that sequential addition of models can progressively reduce error.
             The model assumes that the final prediction is a weighted sum of the predictions from individual weak learners.
             XGBoost assumes that encoded integer values for input variables have an ordinal relationship. It is also robust to missing values, as it               can learn how to handle them implicitly during training. 

**JUSTIFICATION**

The Random Forest is a powerful ensemble method that is highly appropriate for this dataset. It is capable of capturing complex, non-linear relationships that a linear model would likely miss.

Strengths: High accuracy, ability to capture non-linear relationships, robustness to outliers and noisy data, and capability to provide feature importance rankings.

Weaknesses: Can be computationally intensive and memory-hungry(It demands a lot of storage). It is also less interpretable than linear regression.

Assumptions: Assumes that the training data consists of Independent and Identically Distributed observations. For time-series, this is mitigated by the chronological split. 

```python
# 4.2 Model Comparison
# Create comparison table
metrics_df = pd.DataFrame(metrics).T
metrics_df = metrics_df[['MSE', 'RMSE', 'MAE', 'MAPE', 'R²']]
print("\nModel Comparison Table:")
print(metrics_df.round(4))
```

```python
# Visualize model performance (bar charts for metrics)
metrics_df.plot(kind='bar', figsize=(14, 8))
plt.title=('Model Performance Comparison')
plt.ylabel('Metric Value')
plt.xticks(rotation=45)
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Identify the best performing model before tuning (based on highest R²)
best_model = metrics_df['R²'].idxmax()
best_r2 = metrics_df.loc[best_model, 'R²']

print(f"\nBest performing model based on R²: {best_model} with R² = {best_r2:.4f}")
```

##### Based on the performance metrics and visualizations, Best Performing Model (before tuning) is Support Vector Regression (SVR) .

```python
# Hyperparameter
# For XGBoost (assuming it's the best; adjust if not)
if best_model != 'XGBoost':
    print("Assuming XGBoost for demonstration; in practice, tune the best one.")
param_grid = {
    'learning_rate': [0.01, 0.1, 0.3],
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'subsample': [0.6, 0.8, 1.0]
}
```

```python
# Grid Search CV

from sklearn.model_selection import GridSearchCV,TimeSeriesSplit #We have to import Time Series split which ensures that the training data always precedes the validation data, preventing future data from being used to train the model
cv = TimeSeriesSplit(n_splits=3)  # k=3 for small data
grid_search = GridSearchCV(
    estimator=xgb.XGBRegressor(objective='reg:squarederror', random_state=42),
    param_grid=param_grid,
    cv=cv,
    scoring='neg_mean_squared_error',
    verbose=1,
    n_jobs=-1
)
grid_search.fit(X_train_time, y_train_time)
```

```python
# Best parameters
best_params = grid_search.best_params_
print("\nBest hyperparameters after tuning:", best_params)
```

```python
# Tuned model
tuned_model = grid_search.best_estimator_
y_pred_tuned = tuned_model.predict(X_test_time)
```

```python
# Metrics after tuning
mse_tuned = mean_squared_error(y_test_time, y_pred_tuned)
rmse_tuned = np.sqrt(mse_tuned)
mae_tuned = mean_absolute_error(y_test_time, y_pred_tuned)
mape_tuned = mape(y_test_time, y_pred_tuned)
r2_tuned = r2_score(y_test_time, y_pred_tuned)
print(f"\nTuned XGBoost - Test MSE: {mse_tuned:.4f}, RMSE: {rmse_tuned:.4f}, MAE: {mae_tuned:.4f}, MAPE: {mape_tuned:.4f}%, R²: {r2_tuned:.4f}")
```

```python
# Before vs After Comparison
before_metrics = metrics['XGBoost']
after_metrics = {
    'MSE': mse_tuned,
    'RMSE': rmse_tuned,
    'MAE': mae_tuned,
    'MAPE': mape_tuned,
    'R²': r2_tuned
}
```

```python
comparison_df = pd.DataFrame({'Before Tuning': before_metrics, 'After Tuning': after_metrics})
print("\nBefore vs After Tuning Comparison:")
print(comparison_df.round(4))
```

```python
# Improvement
improvement = {}
for metric in ['MSE', 'RMSE', 'MAE', 'MAPE']:
    abs_improve = before_metrics[metric] - after_metrics[metric]
    pct_improve = (abs_improve / before_metrics[metric]) * 100 if before_metrics[metric] != 0 else np.nan
    improvement[metric] = {'Absolute': abs_improve, 'Percentage': pct_improve}
for metric in ['R²']:
    abs_improve = after_metrics[metric] - before_metrics[metric]
    pct_improve = (abs_improve / abs(before_metrics[metric])) * 100 if before_metrics[metric] != 0 else np.nan
    improvement[metric] = {'Absolute': abs_improve, 'Percentage': pct_improve}
```

```python
print("\nImprovement Achieved:")
for metric, imp in improvement.items():
    print(f"{metric}: Absolute = {imp['Absolute']:.4f}, Percentage = {imp['Percentage']:.2f}%")
```

```python
# Visualization of improvement
comparison_df.plot(kind='bar', figsize=(8, 6))
plt.title=('Performance Before vs After Tuning')
plt.ylabel('Metric Value')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Cross-Validation
# Report mean and std from CV
mean_cv_score = -grid_search.best_score_  # Since neg_mse
std_cv_score = grid_search.cv_results_['std_test_score'][grid_search.best_index_]
print(f"\nCross-Validation (k=3) Mean MSE: {mean_cv_score:.4f}, Std Dev: {std_cv_score:.4f}")
```
